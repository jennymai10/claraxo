```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('ux_survey.csv')
df.head()
```

```python
df.dtypes
```

```python
df.columns = [
    'timestamp', 
    'age', 
    'occupation', 
    'satisfaction', 
    'ease_of_navigation', 
    'issues_with_loading', 
    'issues_explanation', 
    'login_functionality', 
    'signup_functionality', 
    'account_setting_functionality', 
    'play_game_functionality', 
    'game_history_functionality', 
    'improvement_suggestion', 
    'new_feature_suggestion', 
    'additional_comments'
]
```

```python
df['occupation'] = df['occupation'].str.strip().str.lower()
df['occupation'].unique()
```

```python
df['age'] = df['age'].astype(int)
df['satisfaction'] = df['satisfaction'].astype(int)
df.dtypes
```

```python
df['ease_of_navigation'] = df['ease_of_navigation'].replace({
    'Very easy': 5,
    'Easy': 4,
    'Neutral': 3,
    'Hard': 2,
    'Very hard': 1
})

df['login_functionality'] = df['login_functionality'].replace({
    'Excellent': 5,
    'Good': 4,
    'Average': 3,
    'Poor': 2,
    'Very Poor': 1
})

df['signup_functionality'] = df['signup_functionality'].replace({
    'Excellent': 5,
    'Good': 4,
    'Average': 3,
    'Poor': 2,
    'Very Poor': 1
})

df['account_setting_functionality'] = df['account_setting_functionality'].replace({
    'Excellent': 5,
    'Good': 4,
    'Average': 3,
    'Poor': 2,
    'Very Poor': 1
})

df['play_game_functionality'] = df['play_game_functionality'].replace({
    'Excellent': 5,
    'Good': 4,
    'Average': 3,
    'Poor': 2,
    'Very Poor': 1
})

df['game_history_functionality'] = df['game_history_functionality'].replace({
    'Excellent': 5,
    'Good': 4,
    'Average': 3,
    'Poor': 2,
    'Very Poor': 1
})
```

```python
df.head()
```

```python
plt.hist(df['satisfaction'], bins=5, range=(1, 5), alpha=0.7, edgecolor='black')
plt.title('Satisfaction Rating Distribution')
plt.xlabel('Satisfaction Rating')
plt.ylabel('Frequency')
plt.xticks([1, 2, 3, 4, 5])
plt.show()
```

```python
plt.hist(df['age'], alpha=0.7, edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
occupation_counts = df['occupation'].value_counts()
occupation_counts.plot(kind='bar', alpha=0.7, edgecolor='black')
plt.title('Occupation Distribution')
plt.xlabel('Occupation')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```

```python
corr = df[['satisfaction', 'ease_of_navigation', 'login_functionality', 'signup_functionality',
           'account_setting_functionality', 'play_game_functionality', 'game_history_functionality']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Heatmap of Features')
plt.show()
```

```python
# Set plot color palette
sns.set_palette('muted')

# Plotting the boxplot with rotated x-axis labels and custom color
plt.figure(figsize=(10, 6))
sns.boxplot(x='occupation', y='satisfaction', data=df, color='lightseagreen')  # "coral blue" shade
plt.title('User Satisfaction Across Different Occupations')
plt.xlabel('Occupation')
plt.ylabel('Satisfaction')

# Rotate x-axis labels
plt.xticks(rotation=45)

plt.show()
```

```python
# Joint Plot of Ease of Navigation and Satisfaction
sns.jointplot(x='ease_of_navigation', y='satisfaction', data=df, kind='reg')
plt.show()
```
